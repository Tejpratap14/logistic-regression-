```python
import statsmodels.api as sm
```

```python
dframe =pd.read_csv('loan.csv')

```

```python
int_rate =dframe['Interest.Rate']
loan_amt =dframe['Loan.Amount']
fico_score = dframe['FICO.Score']
```

```python
y=np.matrix(int_rate).transpose()
```

```python
x1=np.matrix(fico_score).transpose()
x2=np.matrix(loan_amt).transpose()
```

```python
x= np.column_stack([x1,x2])
```

```python
x3=sm.add_constant(x)
```

```python
# OLS takes two array object like dataframe and hels us in prediction
```

```python
model =sm.OLS(y,x3)
```

```python
model_fit =model.fit()
```

```python
print ("the P values are:") ,model_fit.pvalues

print ("The R -squared value is:"), model_fit.rsquared
```
